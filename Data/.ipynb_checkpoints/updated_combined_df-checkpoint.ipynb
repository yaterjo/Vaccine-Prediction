```python
# Imports
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score
from sklearn.metrics import ConfusionMatrixDisplay, confusion_matrix
from sklearn.metrics import roc_auc_score, RocCurveDisplay, roc_curve
```

```python
# Load the datasets
features_df = pd.read_csv('training_set_features.csv')
labels_df = pd.read_csv('training_set_labels.csv')
```

```python
# Summary statistics for numerical features
numerical_summary = features_df.describe()

# Count of unique values for categorical features
categorical_summary = features_df.select_dtypes(include=['object']).nunique()

# Missing values count for features dataset
missing_values_features = features_df.isnull().sum()

# Missing values count for labels dataset
missing_values_labels = labels_df.isnull().sum()

numerical_summary, categorical_summary, missing_values_features, missing_values_labels
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Setting up the figure size and layout
plt.figure(figsize=(20, 15))
plt.subplots_adjust(hspace=0.4, wspace=0.4)

# Plotting distributions of key numerical features
for index, column in enumerate(['h1n1_concern', 'h1n1_knowledge', 'behavioral_antiviral_meds', 'behavioral_wash_hands']):
    plt.subplot(2, 2, index+1)
    sns.countplot(x=column, data=features_df)
    plt.title(f'Distribution of {column}')

plt.show()

# Merging features and labels for correlation analysis
merged_df = features_df.merge(labels_df, on='respondent_id')

# Plotting correlation of features with the target variables
plt.figure(figsize=(12, 6))
sns.heatmap(merged_df[['h1n1_vaccine', 'seasonal_vaccine', 'h1n1_concern', 'h1n1_knowledge', 'behavioral_antiviral_meds', 'behavioral_wash_hands']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation between Features and Target Variables')
plt.show()
```

```python
# Setting up the figure size and layout for box plots
plt.figure(figsize=(18, 10))
plt.subplots_adjust(hspace=0.4, wspace=0.3)

# Selecting features to plot
features_to_plot = ['h1n1_concern', 'h1n1_knowledge', 'behavioral_antiviral_meds', 'behavioral_wash_hands']

# Creating box plots for selected features
for index, feature in enumerate(features_to_plot):
    plt.subplot(2, 2, index+1)
    sns.boxplot(x=features_df[feature])
    plt.title(f'Box Plot of {feature}')

plt.show()
```

```python
pd.crosstab(
    labels_df["h1n1_vaccine"], 
    labels_df["seasonal_vaccine"], 
    margins=True,
    normalize=True
)
```

```python
# Combining datasets on 'respondent_id' and selecting socio-economic variables and 'seasonal_vaccine'
socio_economic_columns = [
    'respondent_id', 'age_group', 'education', 'race', 'sex', 'income_poverty',
    'marital_status', 'rent_or_own', 'employment_status', 'hhs_geo_region',
    'census_msa', 'household_adults', 'household_children', 'seasonal_vaccine'
]

combined_df = merged_df[socio_economic_columns]

# Displaying the first few rows of the combined dataset to verify selection
combined_df.head()
```

```python
# Setting up the figure size and layout for visualizations
plt.figure(figsize=(18, 12))
plt.subplots_adjust(hspace=0.4, wspace=0.3)

# Age group vs. Seasonal Vaccine
plt.subplot(2, 2, 1)
sns.countplot(x='age_group', hue='seasonal_vaccine', data=combined_df)
plt.xticks(rotation=45)
plt.title('Seasonal Vaccine Uptake by Age Group')

# Education vs. Seasonal Vaccine
plt.subplot(2, 2, 2)
sns.countplot(x='education', hue='seasonal_vaccine', data=combined_df)
plt.xticks(rotation=45)
plt.title('Seasonal Vaccine Uptake by Education Level')

# Income Poverty vs. Seasonal Vaccine
plt.subplot(2, 2, 3)
sns.countplot(x='income_poverty', hue='seasonal_vaccine', data=combined_df)
plt.xticks(rotation=45)
plt.title('Seasonal Vaccine Uptake by Income Poverty Level')

# Employment Status vs. Seasonal Vaccine
plt.subplot(2, 2, 4)
sns.countplot(x='employment_status', hue='seasonal_vaccine', data=combined_df)
plt.xticks(rotation=45)
plt.title('Seasonal Vaccine Uptake by Employment Status')

plt.show()
```

```python
# Checking for class imbalances in the 'seasonal_vaccine' target variable
seasonal_vaccine_counts = combined_df['seasonal_vaccine'].value_counts(normalize=True)

seasonal_vaccine_counts
```

```python
# Updating the socio_economic_columns list to exclude 'hhs_geo_region' and 'census_msa'
updated_socio_economic_columns = [
    'respondent_id', 'age_group', 'education', 'race', 'sex', 'income_poverty',
    'marital_status', 'rent_or_own', 'employment_status',
    'household_adults', 'household_children', 'seasonal_vaccine'
]

# Selecting the updated columns from the combined dataset
updated_combined_df = combined_df[updated_socio_economic_columns]

# Displaying the first few rows of the updated dataset to verify the changes
updated_combined_df.head()
```

```python
# Checking for missing values in the updated dataset
missing_values_updated = updated_combined_df.isnull().sum()

# Filtering out columns that have missing values
missing_values_columns = missing_values_updated[missing_values_updated > 0]

missing_values_columns
```

```python
from sklearn.impute import SimpleImputer

# Creating imputers for categorical and numerical variables
categorical_imputer = SimpleImputer(strategy='most_frequent')
numerical_imputer = SimpleImputer(strategy='median')

# Columns to impute
categorical_columns = ['education', 'income_poverty', 'marital_status', 'rent_or_own', 'employment_status']
numerical_columns = ['household_adults', 'household_children']

# Performing imputation
updated_combined_df[categorical_columns] = categorical_imputer.fit_transform(updated_combined_df[categorical_columns])
updated_combined_df[numerical_columns] = numerical_imputer.fit_transform(updated_combined_df[numerical_columns])

# Checking again for missing values to confirm imputation
missing_values_after_imputation = updated_combined_df.isnull().sum()

missing_values_after_imputation
```

```python
from sklearn.preprocessing import StandardScaler

# Initializing the StandardScaler
scaler = StandardScaler()

# Scaling the numerical features
updated_combined_df[numerical_columns] = scaler.fit_transform(updated_combined_df[numerical_columns])

# Displaying the first few rows to verify scaling
updated_combined_df.head()
```

```python
# Creating a new feature 'total_household_size' from 'household_adults' and 'household_children'
updated_combined_df['total_household_size'] = updated_combined_df['household_adults'] + updated_combined_df['household_children']

# Displaying the first few rows to verify the new feature
updated_combined_df[['household_adults', 'household_children', 'total_household_size']].head()
```

```python
# Calculating the correlation matrix for the updated dataset, including the new feature
correlation_matrix = updated_combined_df.corr()

# Displaying the correlation matrix
correlation_matrix
```
