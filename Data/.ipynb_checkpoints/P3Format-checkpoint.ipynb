```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate the number of missing values for each column
missing_values_count = merged_df.isnull().sum()

# Calculate the percentage of missing values for each column
missing_values_percentage = (missing_values_count / len(merged_df)) * 100

# Create a dataframe for missing values
missing_values_df = pd.DataFrame({'column_name': missing_values_count.index, 'missing_values': missing_values_count.values, 'percentage_missing': missing_values_percentage.values})

# Filter out columns that have no missing values to simplify the visualization
missing_values_df = missing_values_df[missing_values_df['missing_values'] > 0]

# Sort dataframe by percentage of missing values
missing_values_df = missing_values_df.sort_values(by='percentage_missing', ascending=False)

# Plotting
plt.figure(figsize=(10, 8))
sns.barplot(x='percentage_missing', y='column_name', data=missing_values_df, palette='viridis')
plt.title('Percentage of Missing Values by Column')
plt.xlabel('Percentage Missing')
plt.ylabel('Column Name')
plt.show()
```

```python
# Identify categorical columns
categorical_cols = merged_df.select_dtypes(include=['object', 'category']).columns.tolist()

# Initialize lists to hold binary and non-binary categorical columns
binary_categorical_cols = []
non_binary_categorical_cols = []

# Separate binary and non-binary categorical columns
for col in categorical_cols:
    if merged_df[col].nunique() == 2:
        binary_categorical_cols.append(col)
    else:
        non_binary_categorical_cols.append(col)

# Prepare to print value counts and percentages for binary and non-binary categorical columns
binary_categorical_values = {}
non_binary_categorical_values = {}

for col in binary_categorical_cols:
    binary_categorical_values[col] = {
        'Value Counts': merged_df[col].value_counts(),
        'Percentage': merged_df[col].value_counts(normalize=True) * 100
    }

for col in non_binary_categorical_cols:
    non_binary_categorical_values[col] = {
        'Value Counts': merged_df[col].value_counts(),
        'Percentage': merged_df[col].value_counts(normalize=True) * 100
    }

binary_categorical_cols, non_binary_categorical_cols
```

```python
# Checking data types of the remaining columns to confirm the approach
data_types = cleaned_df.dtypes

# Identifying columns with missing values
missing_columns = cleaned_df.columns[cleaned_df.isnull().any()].tolist()

# Checking if there are any continuous numerical columns with missing values that are not categorical or binary
continuous_numerical_columns = cleaned_df.select_dtypes(include=['float64', 'int64']).columns.tolist()
continuous_numerical_missing_columns = list(set(continuous_numerical_columns) & set(missing_columns))

# Display data types and identify any continuous numerical columns with missing values
data_types, continuous_numerical_missing_columns
```

```python
# Imputing missing values with the mode
for column in missing_columns:
    mode_value = cleaned_df[column].mode()[0]
    cleaned_df[column].fillna(mode_value, inplace=True)

# Confirming there are no more missing values
missing_values_after_imputation = cleaned_df.isnull().sum().sum()

missing_values_after_imputation
```

```python
from sklearn.preprocessing import StandardScaler

# Selecting numerical columns (including binary and ordinal variables)
numerical_columns = cleaned_df.select_dtypes(include=['int64', 'float64']).columns.tolist()

# Excluding 'seasonal_vaccine' from scaling as it's the target variable
numerical_columns_for_scaling = [col for col in numerical_columns if col != 'seasonal_vaccine']

# Scaling the numerical features
scaler = StandardScaler()
scaled_features = cleaned_df.copy()

scaled_features[numerical_columns_for_scaling] = scaler.fit_transform(cleaned_df[numerical_columns_for_scaling])

# Displaying the first few rows to verify scaling
scaled_features.head()
```

```python
from sklearn.preprocessing import OneHotEncoder

# Identifying categorical columns that need one-hot encoding
# We exclude binary categorical columns as they are already effectively encoded
categorical_columns_to_encode = [col for col in non_binary_categorical_cols if col in cleaned_df.columns]

# One-hot encoding
encoder = OneHotEncoder(sparse=False, drop='first')  # drop='first' to avoid multicollinearity
encoded_features = pd.DataFrame(encoder.fit_transform(cleaned_df[categorical_columns_to_encode]),
                                columns=encoder.get_feature_names_out(categorical_columns_to_encode))

# Dropping original categorical columns from the dataset
cleaned_df_dropped = cleaned_df.drop(columns=categorical_columns_to_encode)

# Concatenating the one-hot encoded features with the rest of the dataset
final_df = pd.concat([cleaned_df_dropped, encoded_features], axis=1)

final_df.head()
```

```python
# Examining class balance for the target variable 'seasonal_vaccine'
class_balance = final_df['seasonal_vaccine'].value_counts(normalize=True) * 100

# Identifying potential outliers using the IQR method for scaled numerical features
# Note: This is more illustrative given that the scaling should have standardized these features
outlier_detection = final_df[numerical_columns_for_scaling].describe()

# Calculating IQR for each feature
Q1 = outlier_detection.loc['25%', :]
Q3 = outlier_detection.loc['75%', :]
IQR = Q3 - Q1

# Defining outliers as those beyond 1.5 * IQR from the Q1 and Q3
outlier_counts = {}
for column in numerical_columns_for_scaling:
    lower_bound = Q1[column] - 1.5 * IQR[column]
    upper_bound = Q3[column] + 1.5 * IQR[column]
    outliers = final_df[(final_df[column] < lower_bound) | (final_df[column] > upper_bound)]
    outlier_counts[column] = len(outliers)

# Converting outlier counts to a DataFrame for easier visualization
outlier_counts_df = pd.DataFrame(list(outlier_counts.items()), columns=['Feature', 'Outliers'])

class_balance, outlier_counts_df.sort_values(by='Outliers', ascending=False)
```

```python
# Preparing the data for training and testing
X = final_df.drop('seasonal_vaccine', axis=1)
y = final_df['seasonal_vaccine']

# Splitting the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Training a Logistic Regression model
model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)

# Making predictions on the test set
y_pred = model.predict(X_test)

# Calculating evaluation metrics
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

# Calculating Specificity from the confusion matrix
tn, fp, fn, tp = confusion_matrix(y_test, y_pred).ravel()
specificity = tn / (tn + fp)

precision, recall, f1, specificity
```
