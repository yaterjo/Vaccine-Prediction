```python
import pandas as pd

# Load the datasets
labels_df = pd.read_csv('training_set_labels.csv')
features_df = pd.read_csv('training_set_features.csv')

# Display the first few rows of each dataset and their summaries
labels_head = labels_df.head()
features_head = features_df.head()

# Check for general information (data types, null values, etc.)
labels_info = labels_df.info()
features_info = features_df.info()

(labels_head, features_head, labels_info, features_info)
```

```python
# Combine the datasets by merging them on the 'respondent_id' column
combined_df = pd.merge(features_df, labels_df, on='respondent_id')

# Display the first few rows of the combined dataset
combined_df_head = combined_df.head()

# Display the shape and basic info of the combined dataset to verify the merge
combined_shape = combined_df.shape
combined_info = combined_df.info()

(combined_df_head, combined_shape, combined_info)
```

```python
# Drop the specified columns, including all that include 'H1N1'
columns_to_drop = [
    'employment_occupation', 'employment_industry', 'health_insurance', 'respondent_id',
    'h1n1_vaccine', 'hhs_geo_region', 'census_msa', 'employment_status'
]

# Add all columns that include 'H1N1' to the list of columns to drop
columns_to_drop.extend([col for col in combined_df.columns if 'h1n1' in col.lower()])

# Drop the specified columns from the combined dataframe
reduced_df = combined_df.drop(columns=columns_to_drop)

# Display the first few rows of the reduced dataframe to confirm
reduced_df_head = reduced_df.head()
reduced_df_head
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the aesthetic style of the plots
sns.set_style("whitegrid")

# Prepare a figure with subplots
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(14, 12))

# Visualization 1: Distribution of Seasonal Vaccine Receipt
sns.countplot(x='seasonal_vaccine', data=reduced_df, ax=axes[0, 0])
axes[0, 0].set_title('Distribution of Seasonal Vaccine Receipt')
axes[0, 0].set_xlabel('Seasonal Vaccine')
axes[0, 0].set_ylabel('Count')
axes[0, 0].set_xticklabels(['Not Received', 'Received'])

# Visualization 2: Age Group Distribution
sns.countplot(y='age_group', data=reduced_df, order=reduced_df['age_group'].value_counts().index, ax=axes[0, 1])
axes[0, 1].set_title('Distribution of Age Groups')
axes[0, 1].set_xlabel('Count')
axes[0, 1].set_ylabel('Age Group')

# Visualization 3: Health Worker Status vs Seasonal Vaccine Receipt
sns.countplot(x='health_worker', hue='seasonal_vaccine', data=reduced_df, ax=axes[1, 0])
axes[1, 0].set_title('Health Worker Status vs Seasonal Vaccine Receipt')
axes[1, 0].set_xlabel('Health Worker Status')
axes[1, 0].set_ylabel('Count')
axes[1, 0].legend(title='Seasonal Vaccine', loc='upper right', labels=['Not Received', 'Received'])

# Visualization 4: Chronic Medical Condition vs Seasonal Vaccine Receipt
sns.countplot(x='chronic_med_condition', hue='seasonal_vaccine', data=reduced_df, ax=axes[1, 1])
axes[1, 1].set_title('Chronic Medical Condition vs Seasonal Vaccine Receipt')
axes[1, 1].set_xlabel('Chronic Medical Condition')
axes[1, 1].set_ylabel('Count')
axes[1, 1].legend(title='Seasonal Vaccine', loc='upper right', labels=['Not Received', 'Received'])

plt.tight_layout()
plt.show()
```

```python
# Address missing values by imputing with the mode for each column
for column in reduced_df.columns:
    # Calculate the mode of each column
    mode_value = reduced_df[column].mode()[0]
    # Fill missing values with the mode
    reduced_df[column].fillna(mode_value, inplace=True)

# Check if there are any remaining missing values
missing_values_after = reduced_df.isnull().sum().sum()

missing_values_after
```

```python
# Correlation Matrix for numerical features
correlation_matrix = reduced_df.corr()

# Plotting the heatmap
plt.figure(figsize=(14, 10))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', cbar=True)
plt.title('Correlation Matrix of Numerical Features')
plt.show()

# For categorical data, we'll look at the relation between some of the categorical variables and the seasonal vaccine uptake
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(14, 12))

# Age Group vs Seasonal Vaccine Uptake
sns.countplot(x='age_group', hue='seasonal_vaccine', data=reduced_df, ax=axes[0, 0], palette='Set2')
axes[0, 0].set_title('Age Group vs Seasonal Vaccine Uptake')
axes[0, 0].tick_params(axis='x', rotation=45)
axes[0, 0].legend(title='Seasonal Vaccine')

# Education Level vs Seasonal Vaccine Uptake
sns.countplot(x='education', hue='seasonal_vaccine', data=reduced_df, ax=axes[0, 1], palette='Set1')
axes[0, 1].set_title('Education Level vs Seasonal Vaccine Uptake')
axes[0, 1].tick_params(axis='x', rotation=45)
axes[0, 1].legend(title='Seasonal Vaccine')

# Income Poverty vs Seasonal Vaccine Uptake
sns.countplot(x='income_poverty', hue='seasonal_vaccine', data=reduced_df, ax=axes[1, 0], palette='Set3')
axes[1, 0].set_title('Income Poverty vs Seasonal Vaccine Uptake')
axes[1, 0].tick_params(axis='x', rotation=45)
axes[1, 0].legend(title='Seasonal Vaccine')

# Sex vs Seasonal Vaccine Uptake
sns.countplot(x='sex', hue='seasonal_vaccine', data=reduced_df, ax=axes[1, 1], palette='coolwarm')
axes[1, 1].set_title('Sex vs Seasonal Vaccine Uptake')
axes[1, 1].legend(title='Seasonal Vaccine')

plt.tight_layout()
plt.show()
```

```python
from sklearn.preprocessing import OneHotEncoder

# Selecting categorical variables
categorical_cols = reduced_df.select_dtypes(include=['object']).columns

# Initializing OneHotEncoder
encoder = OneHotEncoder(sparse=False, drop='first')

# Fitting and transforming the categorical columns
encoded_categorical = encoder.fit_transform(reduced_df[categorical_cols])

# Creating a DataFrame with the encoded variables
encoded_categorical_df = pd.DataFrame(encoded_categorical, columns=encoder.get_feature_names_out(categorical_cols))

# Resetting index on the original DataFrame to concatenate properly
reduced_df.reset_index(drop=True, inplace=True)

# Concatenating the original DataFrame (excluding the original categorical columns) with the new encoded variables DataFrame
encoded_df = pd.concat([reduced_df.drop(columns=categorical_cols), encoded_categorical_df], axis=1)

# Display the first few rows of the newly encoded DataFrame
encoded_df.head()
```

```python
from sklearn.linear_model import LogisticRegression


# Preparing the dataset for the baseline model
X = encoded_df.drop('seasonal_vaccine', axis=1)
y = encoded_df['seasonal_vaccine']

# Creating and training the baseline logistic regression model on the entire dataset
lr_model = LogisticRegression(max_iter=1000)
lr_model.fit(X, y)
```

```python
from sklearn.model_selection import cross_validate
from sklearn.linear_model import LogisticRegression

# Preparing the dataset for the baseline model
X = encoded_df.drop('seasonal_vaccine', axis=1)
y = encoded_df['seasonal_vaccine']

# Creating the logistic regression model
lr_model = LogisticRegression(max_iter=1000, solver='liblinear')

# Configuring cross-validation with desired scoring metrics
scoring = ['accuracy', 'precision', 'recall', 'f1', 'roc_auc']

# Performing cross-validation
cv_results = cross_validate(lr_model, X, y, cv=5, scoring=scoring, return_train_score=False)

# Displaying the average of the evaluation metrics across all cross-validation folds
print(f"Average Accuracy: {cv_results['test_accuracy'].mean():.4f}")
print(f"Average Precision: {cv_results['test_precision'].mean():.4f}")
print(f"Average Recall: {cv_results['test_recall'].mean():.4f}")
print(f"Average F1 Score: {cv_results['test_f1'].mean():.4f}")
print(f"Average ROC-AUC: {cv_results['test_roc_auc'].mean():.4f}")
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score
import numpy as np

# Prepare feature matrix (X) and target vector (y) for the original dataset
X_original = encoded_df.drop('seasonal_vaccine', axis=1)
y_original = encoded_df['seasonal_vaccine']

# Initialize the Decision Tree Classifier model
decision_tree_model = DecisionTreeClassifier(random_state=42)

# Perform 5-fold cross-validation to evaluate model
cv_scores_dt = cross_val_score(decision_tree_model, X_original, y_original, cv=5, scoring='accuracy', n_jobs=-1)

# Calculate the mean and standard deviation of the cross-validation scores
cv_mean_dt = np.mean(cv_scores_dt)
cv_std_dt = np.std(cv_scores_dt)

print(f'Decision Tree - 5-Fold CV Accuracy Mean: {cv_mean_dt:.4f}')
print(f'Decision Tree - 5-Fold CV Accuracy Standard Deviation: {cv_std_dt:.4f}')
```
